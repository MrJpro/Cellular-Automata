# Cellular Automata Rule 184

```python
import cellpylib as cpl

# initialize a CA with 100 cells (a random initialization is also available) 
cellular_automaton = cpl.init_simple(100)

# evolve the CA for 100 time steps, using Rule 184 as defined in NKS
cellular_automaton = cpl.evolve(cellular_automaton, timesteps=100, memoize=True, 
                                apply_rule=lambda n, c, t: cpl.nks_rule(n, 184))

# plot the resulting CA evolution
cpl.plot(cellular_automaton)
```
