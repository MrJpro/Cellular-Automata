# Cellular Automata rule 110

As a linear evolution through time :

```python
import cellpylib as cpl

# initialize a CA with 200 cells (a random initialization is also available) 
cellular_automaton = cpl.init_simple(100)

# evolve the CA for 100 time steps, using Rule 30 as defined in NKS
cellular_automaton = cpl.evolve(cellular_automaton, timesteps=100, memoize=True, 
                                apply_rule=lambda n, c, t: cpl.nks_rule(n, 110))

# plot the resulting CA evolution
cpl.plot(cellular_automaton)
```

As a Two Dimensonal Cellular Automata :

```python
import cellpylib as cpl

# initialize a 60x60 2D cellular automaton
cellular_automaton = cpl.init_simple2d(64, 64)

# evolve the cellular automaton for 30 time steps,
#  applying totalistic rule 126 to each cell with a Moore neighbourhood
for x in range(1,30):
    cellular_automaton = cpl.evolve2d(cellular_automaton, timesteps= x, neighbourhood='Moore',
                                  apply_rule=lambda n, c, t: cpl.totalistic_rule(n, k=2, rule=110))

    cpl.plot2d(cellular_automaton,scale=2.0)
```
